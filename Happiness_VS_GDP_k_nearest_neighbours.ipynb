```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
```

```python
data_root = "https://github.com/ageron/data/raw/main/"
lifesat=pd.read_csv(data_root + "lifesat/lifesat.csv")
lifesat
```

```python
x=lifesat[[ "GDP per capita (USD)" ]].values
y= lifesat[[ "Life satisfaction" ]].values
```

```python
lifesat.plot(kind="scatter", grid="True", x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23500,62500,4,9])
plt.show()
```

```python
model= KNeighborsRegressor(n_neighbors=3) # n_neighbors is the number of nearest neighbors to use for making predictions.
model.fit(x,y)
```

```python
X_new=[[37655.2]]
y_pred=model.predict(X_new)
print(y_pred)
```

## Graph

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), "r-") # plot the regression line
plt.show()
```

```python
lifesat.plot(kind="scatter", grid="True", x="GDP per capita (USD)", y="Life satisfaction")
plt.plot(x,model.predict(x), "r-")
#plt.axis([25000,62500,5.5,7.5])
plt.axis("tight")       #Fits the axes tightly around the data.
plt.plot([X_new[0][0], X_new[0][0]], [0, y_pred[0][0]], "g--")
plt.show()
```
